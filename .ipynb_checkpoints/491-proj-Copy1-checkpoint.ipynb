```python
import numpy as np
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon

# https://data.baltimorecity.gov/Public-Safety/Gun-Offenders/aivj-4x23/data
# https://data.baltimorecity.gov/Public-Safety/Baltimore-Homicides-2018/2prn-qh2n
# https://data.baltimorecity.gov/Housing-Development/Vacant-Buildings/qqcv-ihn5/data
```

```python
def routine(file_name):
    df = pd.read_csv(file_name)
    df = df.drop_duplicates()
    
    def check(loc):
        try:
            float(loc)
            return float(loc)
        except ValueError:
            return np.nan
        
    df['Longitude'] = df['Longitude'].apply(lambda loc: check(loc))
    df['Latitude'] = df['Latitude'].apply(lambda loc: check(loc))
        
    df = df.dropna(subset=['Longitude', 'Latitude'])
    
    df['Coordinates'] = list(zip(df.Longitude, df.Latitude))
    df['Coordinates'] = df['Coordinates'].apply(Point)
    
    return gp.GeoDataFrame(df, geometry='Coordinates')
```

```python
df = pd.read_csv('BPD_Arrests.csv')
df = df.drop_duplicates()

# make sure all the values are floats
def check(loc):
    try:
        float(loc)
        return float(loc)
    except ValueError:
        return np.nan
    
df['Longitude'] = df['Longitude'].apply(lambda loc: check(loc))
df['Latitude'] = df['Latitude'].apply(lambda loc: check(loc))

df = df.dropna(subset=['Longitude', 'Latitude'])
```

```python
# http://geopandas.org/gallery/create_geopandas_from_pandas.html?highlight=geodataframe
df['Coordinates'] = list(zip(df.Longitude, df.Latitude))
df['Coordinates'] = df['Coordinates'].apply(Point)
```
