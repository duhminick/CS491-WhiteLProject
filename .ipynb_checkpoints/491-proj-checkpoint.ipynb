```python
import numpy as np
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
```

```python
df = pd.read_csv('BPD_Arrests.csv')
df = df.drop_duplicates()

# make sure all the values are floats
def check(loc):
    try:
        float(loc)
        return float(loc)
    except ValueError:
        return np.nan
    
df['Longitude'] = df['Longitude'].apply(lambda loc: check(loc))
df['Latitude'] = df['Latitude'].apply(lambda loc: check(loc))

df = df.dropna(subset=['Longitude', 'Latitude'])
```

```python
# http://geopandas.org/gallery/create_geopandas_from_pandas.html?highlight=geodataframe
df['Coordinates'] = list(zip(df.Longitude, df.Latitude))
df['Coordinates'] = df['Coordinates'].apply(Point)
```

```python
# {'init': 'epsg:2248'} from arcgis for previous shapefile
bmore = gp.read_file('./BaltimoreLines/tl_2010_24510_edges.shp')
# unknown coordinate system; it's supposedly epsg:102685; which is used by open baltimore
# however, i think the the openbaltimore website is wrong about this
gdf = gp.GeoDataFrame(df, geometry='Coordinates')

# gdf.crs = '+proj=lcc +lat_1=38.3 +lat_2=39.45 +lat_0=37.66666666666666 +lon_0=-77 +x_0=399999.9999999999 +y_0=0 +ellps=GRS80 +datum=NAD83 +to_meter=0.3048006096012192 no_defs'
# gdf = gdf.to_crs(bmore.crs)
```
